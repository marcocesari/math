## Fibonacci numbers

```python
a = 0
b = 1
print(a)
print(b)
for _ in range(10):
    fib = a+b
    print(fib)
    a = b
    b = fib


```

```python
fibonacci_numbers = [0,1]
for _ in range(10):
    fibonacci_numbers.append(fibonacci_numbers[-1]+fibonacci_numbers[-2])

print(fibonacci_numbers)
```

```python
import matplotlib.pyplot as plt

for i in range(len(fibonacci_numbers)):
    plt.text(i, fibonacci_numbers[i], fibonacci_numbers[i], ha = 'center')
plt.bar(range(len(fibonacci_numbers)), fibonacci_numbers);
```
